Our assumption is that an increase in the number of bathrooms, bedrooms will improve the value of the house; in the same token, we will also assume that the higher the sqft_living or the sqft_lot, the higher the price of the house will be; better grade and better condition will also improve the price of the property We will check our assumptions by applying statsmodels to each of our independant variable versus the dependant variable below.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import statsmodels.stats.api as sms
results = [['ind_var', 'r_squared', 'intercept', 'slope', 'p-value', 'normality (JB)' ]]
for idx, feat in enumerate(['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'condition', 'waterfront', 'view', 'yr_renovated', 'grade', 'lat', 'long', 'zipcode', 'sqft_living15', 'sqft_lot15', 'date_delta']):
     print ("Kc Housing DataSet - Regression Analysis and Diagnostics for formula: price~" + feat)
     print ("-------------------------------------------------------------------------------------")
     f = 'price~' + feat#
     model = smf.ols(formula=f, data=kc_house).fit()
     X_new = pd.DataFrame({feat: [kc_house[feat].min(), kc_house[feat].max()]});
     preds = model.predict(X_new)
     kc_house.plot(kind='scatter', x=feat, y='price');
     plt.plot(X_new, preds, c='red', linewidth=2);
     plt.show()
     fig = plt.figure(figsize=(15,8))
     fig = sm.graphics.plot_regress_exog(model, feat, fig=fig)
     fig = sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True, )
     plt.show()
     results.append([feat, model.rsquared, model.params[0], model.params[1], model.pvalues[1], sms.jarque_bera(model.resid)[0] ])
     input("Press Enter to continue...")# Statsmodel for each variable.
```

```python
pd.DataFrame(results)# parameters of the individual variable via statsmodel
```

Our coefficients and the intercept for the two method are the same, there is a chance our model will generalize well in the future. We will verify that by splitting our data in 2 groups, train and test and compare their mean squared errors.

```python
from sklearn.model_selection import train_test_split# importing the library.
```

```python
linreg.fit(X_train, y_train)
```

```python
y_train_pred = linreg.predict(X_train)
y_test_pred = linreg.predict(X_test)#Calculating predictions on the train set, and on the test set
```

```python
train_residuals = y_train_pred - y_train
test_residuals = y_test_pred - y_test#Calculating the residuals
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
print('Train Mean Squarred Error:', train_mse)
print('Test Mean Squarred Error:', test_mse)# Calculating the mean squared error for test and train set.
```
